#### 1.- Importar librerías para carga de los datasets

```python
import pandas as pd
import numpy as np
```

##### 2.- Carga del dataset y revisión

```python
df = pd.read_csv('hospitalizaciones_train.csv')
```

```python
df.info()
```

```python
# Cuando se trabaja en ML los tipos de datos deben ser númericos (int | float).
```

##### 3.- Verificar si el dataset tiene datos nulos

##### 4.- La columna stay in days muestra la estancia de los pacientes, por definicion se dice que posee estancia hospitalaria prolongada, si ha estado hospitalizado más de 8 días.

```python
df
```

```python
df_ml = df
```

##### 5.- El algoritmo nos dice si tiene estancia prolongada y guarda el valor en la columna predicción.

```python
df_ml["prediccion"] = df_ml["Stay (in days)"].apply(lambda x: 1 if x > 8 else 0)
```

```python
df_ml.tail()
```

##### 6.- Eliminar las columnas que no presentan información relevante para la predicción. 
##### En este caso serían "patientid","Ward_Facility_Code","staff_available","Insurance","Admission_Deposit"

```python
df_ml = df_ml.drop(["patientid","Ward_Facility_Code","staff_available","Insurance","Admission_Deposit"],axis=1)
```

##### 7.- Normalizar los datos para poder realizar ML

```python
df_ml = pd.get_dummies(df_ml)
df_ml
```

```python
df_ml.corr()
```

```python
df_ml = df_ml.drop(['Stay (in days)'], axis = 1)
```

```python
df_ml.info()
```

# Machine Learning
##### Definir 'x' 'y'

```python
X = df_ml.drop(["prediccion"],axis=1)

y = df_ml["prediccion"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=30)
# El parametro 'test_size' sirve para asignar la cantidad de datos en la muestra
```

```python
#standarizar numeros
from sklearn.preprocessing import StandardScaler
escalar = StandardScaler()
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
algorithm = LogisticRegression(max_iter=1000)
```

```python
algorithm.fit(X_train, y_train)
```

##### Realizar la predicción.

```python
y_pred = algorithm.predict(X_test)
```

##### Verificar la matriz de Confusión


```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

##### Calcular la exactitud del modelo

```python
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

##### Calcular la sensibilidad del modelo

```python
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred)
print('Sensibilidad del modelo:')
print(sensibilidad)		
```
